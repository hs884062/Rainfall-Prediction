```python
import pandas as pd
from sklearn.model_selection import train_test_split as tts
```

```python
df=pd.read_csv('weather.csv',low_memory=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Date','Evaporation','Sunshine','Cloud9am','Cloud3pm'])
df.head()
```

```python
for i in df.columns:
    # print(i)
    if df[i].dtypes=='object':
        df[i] = df[i].fillna(df[i].mode()[0])
    else:
        df[i]=df[i].fillna(df[i].median())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.dtypes
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
le4=LabelEncoder()
le5=LabelEncoder()
le6=LabelEncoder()

df['Location']=le1.fit_transform(df['Location'])
df['WindGustDir']=le2.fit_transform(df['WindGustDir'])
df['WindDir9am']=le3.fit_transform(df['WindDir9am'])
df['WindDir3pm']=le4.fit_transform(df['WindDir3pm'])
df['RainToday']=le5.fit_transform(df['RainToday'])
df['RainTomorrow']=le6.fit_transform(df['RainTomorrow'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,Y_train,Y_test=tts(df.drop(columns=['RainTomorrow']),df['RainTomorrow'],test_size=0.3)
```

```python
lr=LogisticRegression(max_iter=10000)
```

```python
lr.fit(X_train,Y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred)*100
```
